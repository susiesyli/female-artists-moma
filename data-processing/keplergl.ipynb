```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df = df.sort_values('acquisition_year')
df
```

```python
df.loc[:, 'timestamp'] = df['acquisition_year'] * 315360000000
```

```python
df
```

## MAKE ARC PAIRS 

```python
sample_points = df.sample(n=20000)[['Longitude', 'Latitude']].values
# Create random pairs
pairs = []
for i in range(8000):
    # Get two random indices
    idx1, idx2 = np.random.choice(len(sample_points), 2, replace=False)
    
    # Get the points
    point1 = sample_points[idx1]
    point2 = sample_points[idx2]
    
    # Add to pairs list
    pairs.append({
        'point1_longitude': point1[0],
        'point1_latitude': point1[1],
        'point2_longitude': point2[0],
        'point2_latitude': point2[1]
    })

# Create the dataframe
pairs_df = pd.DataFrame(pairs)


# Display the result
# print(pairs_df)
```

```python
pairs_artist_df = pairs_df
pairs_artist_df['Artist'] = df['Artist']
pairs_artist_df['Department'] = df['Department']
pairs_artist_df['Nationality'] = df['Nationality']
pairs_artist_df.to_csv('test-arc-artst-3.csv',index=False)
```

```python
def create_geojson_from_df(df):    
    # Create a feature for each point
    features = []
    for _, row in df.iterrows():
        # Convert year to timestamp (multiply by seconds in a year)
        # timestamp = int(row['acquisition_year']) * 31536000
        
        feature = {
            "type": "Feature",
            "properties": {
                "year": str(row['acquisition_year'])
            },
            "geometry": {
                "type": "LineString",  # Changed to Point instead of LineString
                "coordinates": [
                    row['Longitude'],
                    row['Latitude'],
                    0,
                    row['timestamp'] # "timestamp": timestamp
                ]
            },
            # Add timestamp at the feature level
        }
        features.append(feature)
    
    # Create the GeoJSON structure
    geojson = {
        "type": "FeatureCollection",
        "features": features
    }
    
    return geojson
    # coordinates = []
    # for _, row in df.iterrows():
    #     # Convert year to timestamp (multiply by seconds in a year)
    #     coordinates.append([
    #         row['Longitude'],
    #         row['Latitude'],
    #         0,
    #         row['timestamp']
    #     ])

    # # Create the GeoJSON structure
    # geojson = {
    #     "type": "FeatureCollection",
    #     "features": [
    #         {
    #             "type": "Feature",
    #             "properties": {
    #                 "year": str(df['acquisition_year'].iloc[0])  # Using first year as property
    #             },
    #             "geometry": {
    #                 "type": "LineString",
    #                 "coordinates": coordinates
    #             }
    #         }
    #     ]
    # }
    
    # return geojson
```

```python
geojson_data = create_geojson_from_df(df)
```
